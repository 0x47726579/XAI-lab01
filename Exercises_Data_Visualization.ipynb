## Exercise 1 - 

Produce a plot similar to the one shown below.



As shown in this plot, the y values are defined as follows:

- $Y1 = 0$

- $Y2 = (1 - exp(-2 * x)) / (1 + exp(-2 * x))$

- $Y3 = (1 - exp(-0.5 * x)) / (1 + exp(-0.5 * x))$

```python
# The x values
X = np.linspace(-7, 7)
X
```

```python
y1 = [0 for x in X]
#print(len(y1), len(X))
y2 = [ (1- np.exp(-2 * val) )/ (1+ np.exp(-2 * val)) for val in X]
y3 = [ (1- np.exp(-0.5 * val) )/ (1+ np.exp(-0.5 * val)) for val in X]



plt.title('First example with Matplolib')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-7.25,7.25)
plt.ylim(-1.1,1.1)
plt.locator_params(axis="y", nbins=10)
plt.rcParams["figure.figsize"] = (10,5)

plt.plot(X, y1, 'r-.', label="$Y1 = 0$")
plt.plot(X, y2, 'b--', label="$Y2 = (1 - exp(-2 * x)) / (1 + exp(-2 * x))$")
plt.plot(X, y3, 'g', label="$Y3 = (1 - exp(-0.5 * x)) / (1 + exp(-0.5 * x))$")

plt.legend(loc="lower right",fancybox=True)
```

## Exercise 2 - Number of FIFA World Cup wins by teams

Produce a bar chart similar to the one shown below which shows the number of FIFA World Cup wins by teams.



```python
# The data set

FIFA_world_cup_wins = {'Argentina':2, 'Brazil':5, 'England':1, 'France':2, 'Germany':4, 'Italy':4, 'Spain':1, 'Uruguay':2}
```

```python
plt.bar(range(len(FIFA_world_cup_wins)), list(FIFA_world_cup_wins.values()), align='center')
plt.xticks(range(len(FIFA_world_cup_wins)), list(FIFA_world_cup_wins.keys()))

plt.title('FIFA World Cup Winners')
plt.ylabel('Number od FIFA World Cup Wins')
plt.xlabel('National Team')
plt.rcParams["figure.figsize"] = (15,8)

plt.show()
```
